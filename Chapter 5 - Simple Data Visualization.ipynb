# Simple Data Visualization

1. How to chart time series data with line plots and categorical quantities with bar charts
2. How to summarize data distributions with histograms and box plots 
3. How to summarize the relationship between variables with scatter plots

7 parts:
- Data Visualization 
- Introduction to Matplotlib 
- Line Plot
- Bar Chart
- Histogram Plot
- Box and Whisker Plot 
- Scatter Plot

```python
# Basic Matplotlib

# import matplotlib context 
from matplotlib import pyplot 

# create a plot 
pyplot.plot()

# display the plot 
pyplot.show()

# save the plot to file 
pyplot.savefig("myemptyplotimage.png")
```

```python
# Line Plot

# used to present observations collected at regular interval

# create line plot 
# pyplot.plot(x, y)

# example of a line plot 
from numpy import sin 
from matplotlib import pyplot 

# consistent interval for x-axis
x = [x*0.1 for x in range(100)]

# function of x for y-axis
y = sin(x)

# create line plot 
pyplot.plot(x, y)

# show line plot
pyplot.show()
```

```python
# Bar Chart 

# used to present relative quantitoes for multiple categories

# create a bar chart 
# pyplot.bar(x, y)

# example of a bar chart
from random import seed 
from random import randint
from matplotlib import pyplot 

# seed the random number generator 
seed(2)

# names for categories 
x = ['red','green','blue']

# quantities for each categories 
y = [randint(0, 100), randint(0, 100), randint(0, 100)]

# create a bar chart 
pyplot.bar(x, y)

# show line plot 
pyplot.show()
```

```python
# Histogram Plot 

# used to summarize distribution of a data sample 

# create histogram plot 
# pyplot.hist()

# example of a histogram plot 
from numpy.random import seed 
from numpy.random import randn
from matplotlib import pyplot 

# seed the random number generator 
seed(2) 

# random numbers drawn from Gaussian distribution 
x = randn(1000)

# create histogram plot 
pyplot.hist(x) 

# show line plot 
pyplot.show()

# create histogram plot with bins
pyplot.hist(x, bins = 100) 
pyplot.show()
```

```python
# Box and Whisker Plot

# used to summarize the distribution of a data sample, x-axis represent the data sample, where multiple boxplots can be drawn side by side 

# create box and whisker plot
# pyplot.boxplot(x)

# example of a box and a whisker plot 
from numpy.random import seed
from numpy.random import randn
from matplotlib import pyplot 

# seed the random number generator 
seed(1) 

# random numbers drawn from a Gaussian distribution 
x = [randn(1000), 5*randn(1000), 10*randn(1000)]

# create box and whisker plot 
pyplot.boxplot(x) 

# show line plot 
pyplot.show()
```

```python
# Scatter Plot 

# used to summarize the relationship between two paired data samples

# create a scatter plot
# pyplot.scatter(x, y) 

# example of a scatter plot 
from numpy.random import seed 
from numpy.random import randn
from matplotlib import pyplot 

# seed the random number generator
seed(1) 

# first variable 
x = 20*randn(1000) + 100 

# second variable 
y = x + (10 * randn(1000) + 50)

# create scatter plot
pyplot.scatter(x, y) 

# show line plot 
pyplot.show()
```
